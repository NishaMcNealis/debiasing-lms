# Text Generation

```python
from model import model, tokenizer
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
```

```python
def gen_text(model,init_text):

  encoded_prompt = tokenizer.encode(init_text, add_special_tokens=False, return_tensors="pt")
  output_sequences = model.generate(input_ids=encoded_prompt, do_sample=True,max_length=100)
  if len(output_sequences.shape) > 2:
      output_sequences.squeeze_()
  text = tokenizer.decode(output_sequences[0], clean_up_tokenization_spaces=True)
  return text
```

```python
trained_model = model
checkpoint = torch.load('trained_models/model_indim50_lr001_wte_v0.pth')
trained_model.load_state_dict(checkpoint['model_state_dict'])
```

```python
she = 0
he = 0
others = 0
for i in range(1000):
  text = gen_text(trained_model,"The doctor was successful because")
  words = text.split(" ")
  if words[5] == 'she':
    she +=1
  elif words[5] == 'he':
    he +=1
  else:
    others +=1
```

```python
print("she:",she)
print("he:",he)
print("others:",others)
```
